```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
sns.set()
```

```python
data = pd.read_excel("data/iris.csv.xlsx")
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
msno.bar(data, figsize = (6,4))
plt.show()
```

```python
p = data.hist(figsize = (10,8))
```

```python
slm = data.sepal_length.max()
slmi = data.sepal_length.min()
swm = data.sepal_width.max()
swmi = data.sepal_width.min()
plm = data.petal_length.max()
plmi = data.petal_length.min()
pwm = data.petal_width.max()
pwmi = data.petal_width.min()
```

```python
x = ['slm','slmi','swm','swmi','plm','plmi','pwm','pwmi']
y = [slm,slmi,swm,swmi,plm,plmi,pwm,pwmi]
plt.bar(x,y)
plt.title("Max-Min Value Every Columns")
plt.show()
```

## Scatter plot

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=data)
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',data=data)
plt.show()
```

```python
sns.scatterplot(x='sepal_width',y='petal_length',data=data)
plt.show()
```

```python
sns.scatterplot(x='sepal_width',y='petal_width',data=data)
plt.show()
```

## Regplot

```python
sns.regplot(x='sepal_length',y='petal_length',data=data,ci=100)
plt.show()
```

## Distribusion plot All Data

```python
fig, ax = plt.subplots(2,2, figsize=(10,8))
sns.distplot(data.sepal_length, bins = 20, ax=ax[0,0]) 
sns.distplot(data.sepal_width, bins = 20, ax=ax[0,1]) 
sns.distplot(data.petal_length, bins = 20, ax=ax[1,0]) 
sns.distplot(data.petal_width, bins = 20, ax=ax[1,1]) 
plt.show()
```

## Pairplot

```python
plt.figure(figsize=(10,10))
sns.pairplot(data=data)
plt.show()
```

## Violin Plots

```python
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
plt.suptitle('Violin Plots',fontsize=24)
sns.violinplot(x="sepal_length", data=data,ax=ax[0,0],palette='Set3')
sns.violinplot(x="sepal_width", data=data,ax=ax[0,1],palette='Set3')
sns.violinplot (x ='petal_length', data=data, ax=ax[1,0], palette='Set3')
sns.violinplot(x='petal_width', data=data, ax=ax[1,1],palette='Set3')
plt.show()
```

## Box Plot

```python
sns.boxplot(data=data)
plt.show()
```

## Correlation between features

```python
corr=data.corr()

plt.figure(figsize=(14,8))

sns.heatmap(corr, vmax=.8, linewidths=0.01,annot=True,cmap='summer',linecolor="black")
plt.title('Correlation between features')
plt.show()
```
